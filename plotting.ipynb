```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

selected_files = ['result_max_nodes10_max_edges10.csv', 'result_max_nodes10_max_edges30.csv', 'result_max_nodes10_max_edges45.csv']
dfs = {}
for file in selected_files:
    key = file.split('.')[0]
    dfs[key] = pd.read_csv(file)
grouped = {}
for key, df in dfs.items():
    grouped[key] = df.groupby('Num Edges')['Time Taken'].mean()

for key, group in grouped.items():
    print(group)
    plt.plot(group.index, group.values, label=key)

plt.xlabel('Num Edges')
plt.ylabel('Mean Time Taken')
plt.legend()
plt.show()
```

```python
files = ['result_max_nodes10_max_edges45.csv', 'result_max_nodes50_max_edges1225.csv', 'result_max_nodes100_max_edges4950.csv']
dfs = [{}]
for file in files:
    dfs.append(pd.read_csv(file))

# create a line plot of Time Taken vs. Num Edges for each dataframe
for df in dfs:
    plt.plot(df['Num Edges'], df['Time Taken'], label=f'Num Nodes = {df["Num Nodes"][0]}')

# set plot labels and title
plt.xlabel('Num Edges')
plt.ylabel('Time Taken')
plt.title('Progression of Time Taken with Num Nodes')

# add a legend to the plot
plt.legend()

# show the plot
plt.show()
```
